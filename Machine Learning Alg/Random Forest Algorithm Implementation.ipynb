# RandomForestClassifier

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings("ignore")
```

```python
# Train and Test Split Data 
```

```python
tr_features = pd.read_csv('train_features.csv')
tr_labels = pd.read_csv('train_labels.csv', header=None)
```

```python
tr_features.head()
```

```python
tr_features.columns
```

```python
tr_labels.head()
```

```python
# Classifer Models
from sklearn.ensemble import RandomForestClassifier
model=RandomForestClassifier( n_estimators=100,random_state=1).fit(tr_features,tr_labels) # training
```

```python
# Load Test Data 
test_features = pd.read_csv('test_features.csv')
test_labels = pd.read_csv('test_labels.csv', header=None)
```

```python
#Evaluate model
model.score(test_features,test_labels)
```

```python
model.estimators_[99]
```

```python
from sklearn.tree import plot_tree
fig, axes = plt.subplots(nrows = 1,ncols = 1,figsize = (4,4), dpi=800)
plot_tree(model.estimators_[99],feature_names=['Pclass', 'Sex', 'Age', 'Fare', 'Family_cnt', 'Cabin_ind'],filled=True,max_depth=2)
plt.show()
```

```python
new_data=[[2,0,62.0,10.5,0,0]]
model.predict(new_data)
```
